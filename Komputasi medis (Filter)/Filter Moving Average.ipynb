```python
%matplotlib inline
import numpy as np
import pandas as pd
from scipy.io import loadmat
import matplotlib.pyplot as plt
```

```python
def filter_MA(filename,w):
    annots = loadmat(filename)
    data = annots['val'][0]
    data = pd.Series(data,name='heart')
    dataset = data.to_frame()
        
    mov_avg = dataset["heart"].rolling(w).mean()
    mov_mean = mov_avg.astype("float").mean(axis=0)
    dataset['rolling'] = mov_avg.\
    replace(np.nan, mov_mean, inplace=False)
    ploter(dataset["heart"],dataset['rolling'])

def ploter(heart,rolling):            
    plt.title("Filter Digital Moving Average")
    plt.plot(heart, alpha=0.75, color='b', label="noisy signal")
    plt.plot(rolling, color ='g', label="moving average")
    plt.xlabel("Time")
    plt.ylabel("Amplitudo Signal")
    plt.legend(loc=4, framealpha=0.6)
    plt.xlim(0,1500)
    plt.show()
```

```python
#window = 5
filter_MA("212m (26).mat",5)
```

```python
#window = 10
filter_MA("212m (26).mat",10)
```

```python
#window = 20
filter_MA("212m (26).mat",20)
```

```python
#window = 40
filter_MA("212m (26).mat",40)
```
